```python
import torch
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
import requests
import matplotlib.pyplot as plt
import numpy as np
from io import BytesIO
import pandas as pd

model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

final_data_df = pd.read_csv('final_data.csv')
artemis_data_df = pd.read_csv('artemis_dataset_release_v0.csv')

utterance_column = 'utterance'

image_links = final_data_df[['painting', 'Link']].drop_duplicates()

merged_df = artemis_data_df.merge(image_links, left_on='painting', right_on='painting', how='inner')

valid_links = merged_df.groupby('Link').filter(lambda x: len(x) >= 5)['Link'].unique()

sampled_links = np.random.choice(valid_links, 10, replace=False)

for link in sampled_links:
    response = requests.get(link)
    img = Image.open(BytesIO(response.content)).convert("RGB")
    
    utterances = merged_df[merged_df['Link'] == link][utterance_column].tolist()
    
    inputs = processor(text=utterances, images=img, return_tensors="pt", padding=True)

    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.to(device)
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        outputs = model(**inputs)
        logits_per_image = outputs.logits_per_image
        probs = logits_per_image.softmax(dim=1)

    similarities = probs.cpu().numpy()[0]
    for i, utterance in enumerate(utterances):
        print(f"Utterance: {utterance}")
        print(f"Similarity: {similarities[i]:.4f}")
        print()

    max_idx = similarities.argmax()
    predicted_utterance = utterances[max_idx]
    print(f"The predicted utterance is: {predicted_utterance}")

    def visualize_results(image, utterance, similarity):
        fig, ax = plt.subplots()
        ax.imshow(image)
        ax.axis('off')

        plt.title(f"Predicted Utterance:\n{utterance}\nSimilarity: {similarity:.4f}")
        plt.show()

    visualize_results(img, predicted_utterance, similarities[max_idx])
```
